```python
import json
```

```python
with open("fill_in.txt", "r") as f:
    lines = [line.strip() for line in f.readlines() if len(line.strip()) > 0]
lines
```

```python
d = json.load(open("base_template_standart.json", "r"))
```

```python
d
```

```python
dump_lines = []
```

```python
for line in lines:
    splits = line.split(" ")
    path = splits[4][1:-1]
    width = splits[6].split('"')[1]
    style = splits[7].split("=")[-1].split('"')[1]
    name = splits[-1].split("/")[0].split("=")[-1].split('"')[1]
    dump_lines.append( {"impath": path, "width": width, "style": style, "name": name})
d["tree_layout"] = dump_lines
dump_lines
```

```python
json.dump(d, open("base_template_standart.json", "w"), indent=1)
```
